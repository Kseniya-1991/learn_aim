#### Предобработка данных

```python
# Приведем столбцы к нужному типу
nobel['died'] = pd.to_datetime(nobel['died'])
```

```python
nobel['died'].replace('0000-00-00', np.nan, inplace=True)
nobel.head()
```

```python
nobel['born'] = pd.to_datetime(nobel['born'], format='mixed')
```

```python
incomplete_born_df = nobel[nobel['born'].str.contains('-00-00')]
incomplete_born_df.head(3)
```

```python
nobel['incomplete_born'] = np.where(nobel['born'].str.contains('-00-00'), nobel['born'], None)
nobel.head(3)
```

```python
nobel['born'] = nobel['born'].replace(to_replace='.*-00-00', value=np.nan, regex=True)
nobel.head(3)
```

```python
nobel['born'] = pd.to_datetime(nobel['born'], format='mixed')
```

```python
nobel.info()
```

```python
nobel['incomplete_born']
```

```python
nobel['incomplete_born'].str[:4].astype('Int16')
# <NA> - значение для пропущенных данных в целочисленном типе
```

```python
# nobel['born_year'] = nobel['incomplete_born'].str[:4].astype('Int16')
```

```python
nobel.insert(nobel.columns.get_loc('born') + 1, 'born_year', nobel['incomplete_born'].str[:4].astype('Int16'))
nobel.head(3)
```

```python
nobel.shape
```

```python
# Удалим столбец incomplete_born
nobel.drop(columns=['incomplete_born'], inplace=True)
nobel.head(1)
```

```python
nobel['born'].dt.year
```

```python
# Заполним остальные значения в столбце born_year
nobel['born_year'] = nobel['born_year'].fillna(nobel['born'].dt.year)
nobel.head(3)
```

#### Feature Engineering (конструирование признаков)

```python
# nobel['died_year'] = nobel['died'].dt.year
nobel.insert(nobel.columns.get_loc('died') + 1, 'died_year', nobel['died'].dt.year)
nobel.head(3)
```

```python
nobel.tail(3)
```

```python
# Возраст при присуждении премии
nobel['age_at_award'] = nobel['year'] - nobel['born_year']
```

```python
# Продолжительность жизни
current_year = pd.Timestamp.now().year

nobel['years_lived'] = nobel['died_year'] - nobel['born_year']
nobel.fillna({'years_lived': current_year - nobel['born_year']}, inplace=True)

# nobel['years_lived'] = np.where(nobel['died_year'].isna(), current_year - nobel['born_year'], nobel['died_year'] - nobel['born_year'])
```

```python
# Время между присуждением премии и смертью
nobel['years_after_award'] = nobel['died_year'] - nobel['year']
```

```python
# Сколько лет назад была выдана премия
nobel['years_since_award'] = current_year - nobel['year']
```

```python
nobel.head()
```

```python
nobel['decade'] = (np.floor(nobel['year']/10) * 10).astype(int)
```

```python
nobel.tail(3)
```

```python
from wordcloud import WordCloud

# Получаем уникальные категории
categories = nobel['category'].unique()

# Настройка для отображения нескольких облаков слов на одном рисунке
plt.figure(figsize=(15, 10))
rows = (len(categories) + 1) // 2  # определяем количество строк для подграфиков

# Генерация и отображение облака слов для каждой категории
for i, category in enumerate(categories, 1):
    # Фильтруем текст по категории и объединяем все строки `motivation` в одну
    text = " ".join(nobel[nobel['category'] == category]['motivation'].tolist())
    
    # Создаем объект WordCloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    
    # Отображаем облако слов для текущей категории
    plt.subplot(rows, 2, i)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.title(f"Облако слов для категории: {category}")

plt.tight_layout()
plt.show()
```
