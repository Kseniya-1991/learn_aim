```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv',  sep = ',')
display(melb_df.head())
```

```python
exclude_col = ['Date', 'Bedroom', 'Bathroom', 'Rooms', 'Car' ]
max_count = 150

for col in melb_df.columns:
    if melb_df[col].nunique() < max_count and col not in exclude_col:
        melb_df[col] = melb_df[col].astype('category')
display(melb_df.info())
```

```python

display(melb_df.groupby('Regionname')['Lattitude'].std().sort_values(ascending=False))
```

```python
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
date1 = pd.to_datetime('2017-05-01')
date2 = pd.to_datetime('2017-09-01')
mask = (date1 <= melb_df['Date']) & (melb_df['Date']<= date2)
melb_df[mask].groupby('SellerG')['Price'].sum().sort_values(ascending=True)
```

```python
melb_df.columns
```

```python
melb_dfp = melb_df.groupby(['Type', 'Rooms'])['BuildingArea'].median()
display(melb_dfp.unstack())
```

```python
pivot = melb_df.pivot_table(
    values='Price',
    index='SellerG',
    columns='Type',
    aggfunc='median',
)
max_unit_price = pivot['unit'].max()
print(pivot[pivot['unit'] == max_unit_price].index[0])
```

```python
dates = pd.read_csv('data\movies_data\dates.csv', sep= ',')
movies = pd.read_csv('data\movies_data\movies.csv', sep= ',')

ratings1 = pd.read_csv('data\movies_data\\ratings1.csv', sep= ',')
ratings2 = pd.read_csv('data\movies_data\\ratings2.csv', sep= ',')
```

```python

dates['Datetype']= pd.to_datetime(dates['date'])
dates['Datetype'].dt.year.value_counts()
```

```python
ratings = pd.concat([ratings1, ratings2], ignore_index=True)
ratings
```

```python
print('Число строк в таблице ratings: ', ratings.shape[0])
print('Число строк в таблице dates: ', dates.shape[0])
print(ratings.shape[0] == dates.shape[0])
```

```python
display(ratings1.tail(1))
display(ratings2.head(1))
```

```python
ratings = ratings.drop_duplicates(ignore_index=True)
print('Число строк в таблице ratings: ', ratings.shape[0])
# Число строк в таблице ratings: 100836
```

```python
dates.head()
```

```python
ratings_dates = pd.concat([ratings, dates], axis=1)
display(ratings_dates.tail(7))
```

```python
joined_false = ratings_dates.join(
    movies,
    rsuffix='_right',
    how='left'
)
display(joined_false)
```

```python
joined = ratings_dates.join(
    movies.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined.head())
```

```python
merged = ratings_dates.merge(
    movies,
    on='movieId',
    how='left'
)
display(merged.head())
```
